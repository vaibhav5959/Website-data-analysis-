```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/data-export (1).csv")
```

```
df.head()
```

```
df.columns = df.iloc[0]
df.head()
```

```
df = df.drop(0).reset_index(drop=True)
df.head()
```

```
df.dtypes
```

```
df.columns = ['Channel','DateHour','Users','Sessions','Engaged sessions','Average engagement time per session','Engaged sessions per user','Events per session','Engagement rate','Event count']
df.head()
```

```
df['DateHour'] = pd.to_datetime(df['DateHour'],format='%Y%m%d%H',errors='coerce')
```

```
numeric_cols = df.drop(['Channel','DateHour'],axis =1)
df[numeric_cols.columns] = numeric_cols.apply(pd.to_numeric,errors='coerce')
```

```
df["Hour"] = df["DateHour"].dt.hour
df.head()
```

```
# what patterns or trends can you observe in website sessions and users over time?
structured = df.groupby(['DateHour'])[['Sessions','Users']].sum()
figure = plt.figure(figsize=(10,6))
plt.xticks(rotation=90)
plt.title('Sessions and Users Over Time')
plt.xlabel('Date')
plt.ylabel('Sessions and Users')
plt.plot(structured)
```

```
# Which marketing channel brought the highest number of users to the website , and how can we use this insight to imporve traffic from other sources ?
plt.figure(figsize=(10,6))
sns.barplot(x='Channel',y = 'Users',data= df,palette='viridis',estimator=np.sum)
plt.title("Users by channel")
```

```
# Which channel has the highest average engagement time , and what does that tell us about user behavior and content effectiveness ?
plt.figure(figsize=(16,6))
sns.barplot(x= "Channel",y= "Average engagement time per session",palette='viridis',estimator=np.mean,data = df)
plt.title("Average engagement time per session by channel")
plt.grid()
```

```
# How does engagement rate vary across different traffic channels ?
plt.figure(figsize=(16,6))
sns.boxplot(x= "Channel",y= "Engagement rate",palette='viridis',data = df)
plt.title("Engagement rate by channel")
plt.grid()
```

```
# Which channels are driving more engaged sessions compared to non-engaged ones , and what strategies can improve engagement in underperforming channels?
sessions_df = df.groupby('Channel')[['Sessions','Engaged sessions']].sum().reset_index()
sessions_df['non_engaged']= sessions_df['Sessions'] - sessions_df['Engaged sessions']
sessions_df_melted = sessions_df.melt(id_vars='Channel',value_vars=['Engaged sessions','non_engaged'])
plt.figure(figsize=(8,5))
sns.barplot(data=sessions_df_melted, x="Channel",y= "value",hue="variable")
plt.title("Engaged vs non engaged sessions")
plt.show()
```

```
# At what hours of the day does each channel drive the most traffic ?
traffic = df.groupby(['Channel','Hour'])['Sessions'].sum().unstack().fillna(0)
plt.figure(figsize=(20,6))
sns.heatmap(traffic ,annot=True, fmt = '.0f')
plt.title("Hour bs Channel throughout the day")
plt.show()
```

```
# Is there any correlation between high traffic and high engagement rate over time ?
df1 = df.groupby(['DateHour'])[['Sessions','Engaged sessions','Engagement rate']].mean().reset_index()
plt.figure(figsize=(16,6))
plt.plot(df1['DateHour'],df1['Sessions'],label="Sessions",color='red')
plt.plot(df1['DateHour'],df1['Engagement rate'],label="Engagement rate",color='blue')
plt.xlabel='date hour'
plt.title('Sessions and Engagement rate over time')
plt.legend()
plt.grid()
plt.show()
```
